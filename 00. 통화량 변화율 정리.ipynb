```python
import pandas as pd
```

```python
monetary=pd.read_csv('data/monetary.csv', encoding='cp949')
```

```python
monetary.head()
```

```python
data=monetary.loc[:1,].T
data=data[4:]
data.columns = ['korea monetary', 'us monetary']
```

```python
data
```

```python
korea_rate=data['korea monetary']
us_rate=data['us monetary']
```

```python
korea_rate_1=[]
for i in range(len(korea_rate)-1):
    rate=round((float(korea_rate[i+1]/korea_rate[i])-1)*100,2)
    korea_rate_1.append(rate)
```

```python
korea_rate_1
```

```python
us_rate_1=[]
for i in range(len(us_rate)-1):
    rate=round((float(us_rate[i+1]/us_rate[i])-1)*100,2)
    us_rate_1.append(rate)
```

```python
us_rate_1
```

```python
data_1=data[1:]
data_1['korea_rate']=korea_rate_1
data_1['us_rate']=us_rate_1
```
